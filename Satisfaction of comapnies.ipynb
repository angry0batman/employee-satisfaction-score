```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("Analyser dataset.csv")
```

```python
data.head(10)
```

```python
mask1=data["company"]=="Reliance"
mask1data=data[mask1]
mask2=data["company"]=="Tata"
mask2data=data[mask2]
mask3=data["company"]=="Adani"
mask3data=data[mask3]
mask4=data["company"]=="infosys"
mask4data=data[mask4]
mask5=data["company"]=="wipro"
mask5data=data[mask5]
```

```python
print("Number of employees in Reliance:",len(mask1data))
print("Number of employees in Tata:",len(mask2data))
print("Number of employees in Adani:",len(mask3data))
print("Number of employees in infosys:",len(mask4data))
print("Number of employees in wipro:",len(mask5data))
```

```python
import matplotlib.pyplot as plt
```

```python
x = ["Reliance","Tata","Adani","infosys","wipro"]
y = [len(mask1data),len(mask2data),len(mask3data),len(mask4data),len(mask5data)]


plt.bar(x, y)
plt.xlabel('company name')
plt.ylabel('No of employees')
plt.title('company and their number of employees')
plt.show()
```

```python
workinghours_mean1 = mask1data["average_montly_hours"].mean()
workinghours_mean2 = mask2data["average_montly_hours"].mean()
workinghours_mean3 = mask3data["average_montly_hours"].mean()
workinghours_mean4 = mask4data["average_montly_hours"].mean()
workinghours_mean5 = mask5data["average_montly_hours"].mean()


print("Mean of the working hours for Reliance:", workinghours_mean1)
print("Mean of the working hours for Tata:", workinghours_mean2)
print("Mean of the working hours for Adani:", workinghours_mean3)
print("Mean of the working hours for infosys:", workinghours_mean4)
print("Mean of the working hours for wipro:", workinghours_mean5)
```

```python
x = ["Reliance","Tata","Adani","infosys","wipro"]
y = [workinghours_mean1,workinghours_mean2,workinghours_mean3,workinghours_mean4,workinghours_mean5]


plt.plot(x, y)
plt.xlabel('company name')
plt.ylabel('average working hours')
plt.title('company and their average working hours')
plt.show()
```

```python
project_mean1 = mask1data["number_project"].mean()
project_mean2 = mask2data["number_project"].mean()
project_mean3 = mask3data["number_project"].mean()
project_mean4 = mask4data["number_project"].mean()
project_mean5 = mask5data["number_project"].mean()


print("Mean of the projects for Reliance:",project_mean1 )
print("Mean of the  for Tata:", project_mean2)
print("Mean of the projects for Adani:", project_mean3)
print("Mean of the projects for infosys:", project_mean4)
print("Mean of the projects for wipro:", project_mean5)
```

```python
x = ["Reliance","Tata","Adani","infosys","wipro"]
y = [project_mean1,project_mean2,project_mean3,project_mean4,project_mean5]


plt.plot(x, y)
plt.xlabel('company name')
plt.ylabel('average projects')
plt.title('company and their average project')
plt.show()
```

```python


prom1 = mask1data["promotion_last_5years"] == 1
prom2 = mask2data["promotion_last_5years"] == 1
prom3 = mask3data["promotion_last_5years"] == 1
prom4 = mask4data["promotion_last_5years"] == 1
prom5 = mask5data["promotion_last_5years"] == 1


prom_count1 = len(mask1data[prom1])
prom_count2 = len(mask2data[prom2])
prom_count3 = len(mask3data[prom3])
prom_count4 = len(mask4data[prom4])
prom_count5 = len(mask5data[prom5])

print("Promotions in Reliance:", prom_count1)
print("Promotions in Tata:", prom_count2)
print("Promotions in Adani:", prom_count3)
print("Promotions in Infosys:", prom_count4)
print("Promotions in Wipro:", prom_count5)
print("Total Promotions:",prom_count1+prom_count2+prom_count3+prom_count4+prom_count5)
```

```python
x = ["Reliance","Tata","Adani","infosys","wipro"]
y = [prom_count1,prom_count2,prom_count3,prom_count4,prom_count5]


plt.bar(x, y)
plt.xlabel('company name')
plt.ylabel('No of employee getting promotions')
plt.title('company and their number of employees who got promotions in last 5 years')
plt.show()
```
